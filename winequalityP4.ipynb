```python
import pandas as pd

# 1. Membaca dataset
df = pd.read_csv('winequality-red.csv')

# 2. Menghitung rata-rata nilai 'quality' berdasarkan kategori low dan high tiap fitur
kolom = ['alcohol', 'pH', 'residual sugar', 'citric acid']

for col in kolom:
    # Membagi data menjadi 2 rentang (low dan high)
    df[col + '_kategori'] = pd.cut(
        df[col],
        bins=2,
        labels=['low', 'high']
    )

    # Menghitung rata-rata skor quality
    hasil = df.groupby(
        col + '_kategori',
        observed=False
    )['quality'].mean()

    print(f"--- Fitur: {col} ---")
    print(hasil)
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram Distribusi Quality
plt.figure(figsize=(6,4))
sns.histplot(df['quality'], kde=True, color='skyblue')
plt.title('Distribusi Quality')
plt.xlabel('Skor Quality')
plt.ylabel('Frekuensi')
plt.show()

# 2. Heatmap Korelasi Antar Fitur
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasi Antar Fitur')
plt.show()

# 3. Boxplot Alcohol terhadap Quality
plt.figure(figsize=(6,4))
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set2')
plt.title('Alcohol vs Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()

# 4. Scatterplot Citric Acid dan Alcohol terhadap Quality
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='alcohol',
    y='citric acid',
    hue='quality',
    data=df,
    palette='viridis'
)
plt.title('Alcohol dan Citric Acid terhadap Quality')
plt.xlabel('Alcohol')
plt.ylabel('Citric Acid')
plt.show()
```
